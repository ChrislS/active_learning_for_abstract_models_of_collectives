# Uncertainty sampling, guided by the `fertilized forest` library

```python
# Imports
import numpy as np
np.random.seed(0)
import matplotlib.pylab as plt
import scipy.stats
import fertilized
import GPy
from scipy import signal
from PLF import PLF
from IPython import display
import time
from sklearn.metrics import mean_squared_error
```

```python
# Read the data and prepare it for later use.
path_to_csv = "Costs50.csv"
data = np.genfromtxt(path_to_csv, dtype=float, delimiter=';')
power50 = data[:,0]
costs50 = data[:,1]
# For training.
X = np.ascontiguousarray(np.atleast_2d(power50).T)
Y = np.ascontiguousarray(np.atleast_2d(costs50).T)
```

```python
# The marker parameters for plotting.
markerparams = {'marker':'o',
                'markerfacecolor':'None',
                'markeredgecolor':'blue',
                'markersize':6,
                'markeredgewidth':1}
```

```python
plt.plot(power50, costs50, **markerparams)
plt.xlabel('$P$')
plt.ylabel('Cost')
```

Use double as datatype, configure the forest for regression.

```python
soil = fertilized.Soil('d', 'd', 'd', fertilized.Result_Types.regression)
forest = soil.StandardRegressionForest(1, # Number of features
                                       3, # Maximum depth
                                       1, # Tested features per node
                                       1, # Tested thresholds
                                       32, # # trees
                                       7, # Min samples per leaf
                                       14,  # Min samples per split
                                       min_gain_threshold=0., # Gain threshold
                                       summary_mode=1) # One summarizing gaussian
forest.fit(X, Y)
```

Plot the interval [0, 600] with 600 samples.

```python
def create_plot(forest, title_add='', marked_points=None, plot_grid=(1,1)):
    # Define the x values for prediction input.
    X_test = np.linspace(0, np.max(X), num=600)
    X_test = np.atleast_2d(X_test).T
    # Predict the y values based on X_test
    prediction = forest.predict(X_test)
    y_upper_conf = []
    y_lower_conf = []
    for x_ind, x in enumerate(X_test):
        try:
            interval = scipy.stats.norm.interval(0.95, loc=prediction[x_ind, 0], scale=np.sqrt(prediction[x_ind, 1]))
        except:
            interval = scipy.stats.norm.interval(0.95, loc=prediction[0][x_ind, 0], scale=np.sqrt(prediction[1][x_ind]))
        y_lower_conf.append(interval[0])
        y_upper_conf.append(interval[1])
    # Create the plot.
    plt.subplot(plot_grid[0], plot_grid[1], 1)
    plt.title(r'Data and uncertainty ' + title_add)
    plt.plot(X, Y, label='Observations', **markerparams)
    if not marked_points is None:
        plt.plot(X[marked_points], Y[marked_points], label='Used points',
                 marker='x', markeredgecolor='red', markeredgewidth=1,
                 linestyle='None')
    try:
        conv_var = np.convolve(prediction[:, 1], signal.gaussian(50, 9.), mode='same')
    except:
        conv_var = np.convolve(prediction[1][:, 0], signal.gaussian(50, 10.), mode='same')
    plt.plot(X_test, np.sqrt(conv_var), label=r'$\sigma$ convolved')
    plt.xlabel('$P$')
    plt.ylabel('Cost / $\sigma$ convolved')
    plt.legend(loc='upper left')
plt.figure(figsize=(15, 5))
create_plot(forest, marked_points=range(X.shape[0]))
```

## Active learning

```python
def showAndRedraw(delay):
    r"""
    Displays the current plot and waits `delay` seconds
    until continuing.
    """
    display.clear_output()
    display.display(plt.gcf())
    time.sleep(delay)
    plt.gcf().clear()
```

```python
plt.figure(figsize=(15, 10))
sample_points = np.round(np.linspace(0, power50.shape[0]-1, num=12)).astype('int')
errors_guided_plf = []
errors_sample_counts_guided_plf = []
for add_point_idx in xrange(X.shape[0] - 11):
    forest = soil.StandardRegressionForest(1, # Number of features
                                           3, # Maximum depth
                                           1, # Tested features per node
                                           1, # Tested thresholds
                                           32, # # trees
                                           7, # Min samples per leaf
                                           14,  # Min samples per split
                                           min_gain_threshold=0., # Gain threshold
                                           summary_mode=1) # One summarizing gaussian
    plf_model = PLF()
    X_sampled = X[sample_points, :]
    Y_sampled = Y[sample_points, :]
    forest.fit(X_sampled, Y_sampled)
    plf_model.fit(X_sampled[:, 0], Y_sampled[:, 0])
    create_plot(forest,
                ' using %d samples (%d total)' % (np.unique(sample_points).shape[0],
                                                  sample_points.shape[0]),
                sample_points,
                (2, 1))
    # Measure the error.
    error_guided_plf = mean_squared_error(Y, plf_model.predict(X))
    errors_sample_counts_guided_plf.append(np.unique(sample_points).shape[0])
    errors_guided_plf.append(error_guided_plf)
    plt.subplot(2,1,2)
    plt.title('Error using %d samples (%d total)' % (np.unique(sample_points).shape[0],
                                                     sample_points.shape[0]))
    plt.xlabel('Number of sample points')
    plt.ylabel('MSE')
    plt.plot(errors_sample_counts_guided_plf, errors_guided_plf, label='MSE guided PLF')
    plt.xlim((12, power50.shape[0]))
    plt.ylim((0., 2.5e9))
    plt.legend(loc='upper left')
    showAndRedraw(0.2)
    # Find the most promising point and resample.
    X_test = np.linspace(0, np.max(X), num=600)
    X_test = np.atleast_2d(X_test).T
    prediction = forest.predict(X_test)
    suggested_x = None
    conv_var = np.convolve(prediction[:, 1], signal.gaussian(100, 9.), mode='same')
    candidates = range(prediction.shape[0])
    if len(sample_points) < X.shape[0]:
        while suggested_x is None:
            suggested_x_idx = np.argmax(conv_var[candidates])
            suggested_x_idx = candidates[suggested_x_idx]
            suggested_x = X_test[suggested_x_idx]
            # Get closest known powervalue.
            closest_x_idx = np.argmin((power50-suggested_x)**2)
            if closest_x_idx in sample_points:
                candidates = [candidate for candidate in candidates if candidate != suggested_x_idx]
                suggested_x = None
        sample_points = np.hstack((sample_points, closest_x_idx))
```

## Direct evaluation of the piecewise linear functions

```python
plt.figure(figsize=(15, 10))
errors_plf = []
errors_sample_counts_plf = []
plf_sample_points = 12
for add_point_idx in xrange(X.shape[0]-11):
    sample_points = np.round(np.linspace(0, power50.shape[0]-1, num=plf_sample_points)).astype('int')
    plf_model = PLF()
    X_sampled = X[sample_points, 0]
    Y_sampled = Y[sample_points, 0]
    plf_model.fit(X_sampled, Y_sampled)
    plt.subplot(2, 1, 1)
    # Define the x values for prediction input.
    X_test = np.linspace(0, np.max(X), num=600)
    X_test = np.atleast_2d(X_test).T
    # Predict the y values based on X_test
    prediction = plf_model.predict(X_test)
    plt.title('Data and prediction for %d points' % (12 + add_point_idx))
    plt.plot(X, Y, label='Observations', linestyle='None', **markerparams)
    plt.plot(X_test, prediction, label='Prediction')
    # Measure the error.
    error = mean_squared_error(Y, plf_model.predict(X)[:, 0])
    errors_sample_counts_plf.append(np.unique(sample_points).shape[0])
    errors_plf.append(error)
    plt.subplot(2,1,2)
    plt.title('Error using %d samples (%d total)' % (np.unique(sample_points).shape[0],
                                                     sample_points.shape[0]))
    plt.xlabel('Number of sample points')
    plt.ylabel('MSE')
    plt.plot(errors_sample_counts_guided_plf, errors_guided_plf, label='MSE guided PLF')
    plt.plot(errors_sample_counts_plf, errors_plf, label='MSE pure PLF')
    plt.legend(loc='upper right')
    showAndRedraw(0.2)
    plf_sample_points += 1
```

## Gaussian processes

```python
plt.figure(figsize=(15, 10))
sample_points = np.round(np.linspace(0, power50.shape[0]-1, num=12)).astype('int')
errors_guidedgp_plf = []
errors_sample_counts_guidedgp_plf = []
for add_point_idx in xrange(X.shape[0] - 11):
    X_sampled = X[sample_points, :]
    Y_sampled = Y[sample_points, :]
    model = GPy.models.GPRegression(X_sampled, Y_sampled, GPy.kern.Brownian(1, variance=75.))
    plf_model = PLF()
    plf_model.fit(X_sampled[:, 0], Y_sampled[:, 0])
    model.optimize_restarts()
    create_plot(model,
                ' using %d samples (%d total)' % (np.unique(sample_points).shape[0],
                                                  sample_points.shape[0]),
                sample_points,
                (2, 1))
    # Measure the error.
    error_guided_plf = mean_squared_error(Y, plf_model.predict(X))
    errors_sample_counts_guidedgp_plf.append(np.unique(sample_points).shape[0])
    errors_guidedgp_plf.append(error_guided_plf)
    plt.subplot(2,1,2)
    plt.title('Error using %d samples (%d total)' % (np.unique(sample_points).shape[0],
                                                     sample_points.shape[0]))
    plt.xlabel('Number of sample points')
    plt.ylabel('MSE')
    plt.plot(errors_sample_counts_guided_plf, errors_guided_plf, label='MSE DF guided PLF')
    plt.plot(errors_sample_counts_plf, errors_plf, label='MSE pure PLF')
    plt.plot(errors_sample_counts_guidedgp_plf, errors_guidedgp_plf, label='MSE GP guided PLF')
    plt.xlim((12, power50.shape[0]))
    plt.ylim((0., 2.5e9))
    plt.legend(loc='upper right')
    showAndRedraw(0.2)
    # Find the most promising point and resample.
    X_test = np.linspace(0, np.max(X), num=600)
    X_test = np.atleast_2d(X_test).T
    prediction = model.predict(X_test)
    suggested_x = None
    conv_var = np.convolve(prediction[1][:, 0], signal.gaussian(100, 9.), mode='same')
    candidates = range(prediction[0].shape[0])
    if len(sample_points) < X.shape[0]:
        while suggested_x is None:
            suggested_x_idx = np.argmax(conv_var[candidates])
            suggested_x_idx = candidates[suggested_x_idx]
            suggested_x = X_test[suggested_x_idx]
            # Get closest known powervalue.
            closest_x_idx = np.argmin((power50-suggested_x)**2)
            if closest_x_idx in sample_points:
                candidates = [candidate for candidate in candidates if candidate != suggested_x_idx]
                suggested_x = None
        sample_points = np.hstack((sample_points, closest_x_idx))
```
